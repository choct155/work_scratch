# The [Jupyter](http://jupyter.org/) Notebook

```python
import numpy as np
import pandas as pd
from pandas import Series, DataFrame

from IPython.display import HTML
from IPython.display import Image

import ipywidgets as widgets
from ipywidgets import interact, interactive, fixed
```

The Jupyter project provides a fairly language-agnostic analytic platform:

1. It currently supports over [40 languages](https://github.com/ipython/ipython/wiki/IPython-kernels-for-other-languages);
2. It easily shared (including HTML conversions for folks without the platform);
3. It features interactive [widgets](https://github.com/ipython/ipywidgets/blob/master/examples/notebooks/Widget%20Basics.ipynb);
4. It can capture a variety of media;
5. It can interact with the shell; and,
6. It provides instant documentation.

## Sharing

The Notebooks themselves are fairly lightweight [JSON](http://www.json.org/) files.  They can be opened and edited by anyone who has Jupyter Notebook (and the language-appropriate kernel).  The real gem for static formats, however, is [nbconvert](http://nbconvert.readthedocs.org/en/latest/).  This functionality can convert dynamic Notebooks into static files in a number of formats, including the following:

1. Standalone HTML and Reveal.js slide decks (which can be opened by anyone with a browser);
2. LaTeX
3. PDF
