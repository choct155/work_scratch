# Calculating the Gini Coefficient

To better understand the implications of imputing Medicaid receipt for non-reporters in the CPS, it is useful to examine the implications for inequality in income across the population.  Presumably, the cash value of Medicaid serves to increase real income and thereby reduce income inequality.  To test this proposition, it would be useful to have a pre-built, efficient function we could use to calculate the Gini coefficient.  Developing this function is the purpose of this Notebook.  That being said, we could certainly add functions for other inequality measures down the road.

## Gini Coefficient

The Gini coefficient is a function of the Lorenz Curve, which is just a plot of the cumulative share of income versus the cumulative share of people (ranked from lowest to highest income).

```python
Image('http://mathscinotes.files.wordpress.com/2011/10/incomeinequity.png')
```

Intuitively, the Gini coefficient is the deviation from perfect equality of income.  If everyone has the same income, the percentage of income and the population share increase 1-to-1, thereby resulting in the 45 degree angle in the image above.  As inequality increases, the Lorenz curve moves off the diagonal and becomes more concave.  The extent of this deviation is captured by the area of `A`.  The fraction of the total area under the diagonal that is taken by `A` is our measure.

Amartya Sen (*On Economic Inequality, 1977*) demonstrated that the geometric expression above is equivalent to half the relative mean absolute difference.  Let's unpack that a bit.  The absolute difference between any two members of the population is straighforward.

$$\text{abs_diff} = |x_i - x_j|$$

The mean absolute difference takes the average of `abs_diff` for all pairs.

$$\text{mean_abs_diff} = \frac{\sum_i \sum_j |x_i - x_j|}{n^2}$$

Note that we are dividing by $n^2$ because we are comparing each of $n$ values with all $n$ values in the data set.  So, we want the average of $n^2$ relations.  The relative mean absolute difference just normalizes the mean absolute difference via dividing it by the average income of the population.

$$\text{rel_mean_abs_diff} = \frac{\sum_i \sum_j |x_i - x_j|}{n^2} \frac{n}{\sum_i x_i} = \frac{\sum_i \sum_j |x_i - x_j|}{n^2 \bar x} = 2G$$

Therefore, $G$ can be calculated in a straightforward way.

$$G = \frac{\sum_i \sum_j |x_i - x_j|}{2n^2 \bar x}$$

This is not the most efficient way to calculate this measure, but it is very robust.  More efficient methods generally require preprocessing or certain distributional conditions.  Let's generate a function to calculate this measure.

```python
def gini(inc_raw):
    '''Function takes an array of incomes. It returns the Gini cofficient for the income distribution.'''
    #Capture normalized income array
    inc_arr=inc_raw/float(inc_raw.sum())
    #Capture length of array
    n=len(inc_arr)
    #Capture average income
    avg_inc=inc_arr.mean()
    #Initialize sum of absolute difference
    abs_diff_sum=0
    #For each unit...
    for i in range(n):
        #...take every other unit...
        for j in range(n):
            #...and update the sum of absolute differences with the comparison
            abs_diff_sum+=abs(inc_arr[i]-inc_arr[j])
    #Calculate the Gini coefficient
    g=abs_diff_sum/(2*avg_inc*n**2)

    return g
    
gini(np.random.choice(np.arange(1000)*10,size=1000))  
```

It appears to work.  Isn't that nice?  Too bad we have a wrinkle.  Since we are using survey data, we must incorporate weights!  To make this happen, we replace the sum of absolute differences in the numerator with the weighted sum of absolute differences...

$$\sum_i \sum_j |x_i - x_j| \rightarrow \sum_i w_i \sum_j w_j|x_i - x_j|$$

...and replace the population count in the denominator ($n$) with the weighted population count ($\sum_i^n w_i$).  The full weighted calculation is as follows:

$$G_w = \frac{\sum_i w_i \sum_j w_j|x_i - x_j|}{2(\sum_i^n w_i)^2 \bar x}$$

(Hat tip to [Dr. John Creedy](http://www.victoria.ac.nz/sacl/about/cpf/publications/pdfs/2015-pubs/WP03_2015_Gini_Inequality.pdf).)

Let's go ahead and generate a new function for this weighted version.

```python
def gini_wt(inc_raw,wt_raw=1):
    '''Function takes an array of incomes and sample weights. It returns the Gini cofficient for the income distribution.'''
    #If no weight array has been provided, assume equal weights
    if isinstance(wt_raw,int):
        wt_raw=np.ones(len(inc_raw))
    
    #Confirm we have weights for all incomes and vice versa
    try:
        assert len(inc_raw)==len(wt_raw)
    except:
        print 'The length of the income and weight arrays do not match.  Womp womp...'
        #Return null value for Gini
        g=np.nan
        return g
    
    #Capture normalized income array
    inc_arr=inc_raw/float(inc_raw.sum())
    #Capture normalized weight array
    wt_arr=wt_raw/float(wt_raw.sum())
    #Capture length of array
    n=len(inc_arr)
    #Capture unweighted average income
    avg_inc=inc_arr.mean()
    #Initialize sum of absolute difference
    abs_diff_sum=0
    #For each unit...
    for i in range(n):
        #...take every other unit...
        for j in range(n):
            #...and update the sum of absolute differences with the comparison
            abs_diff_sum+=wt_arr[i]*wt_arr[j]*abs(inc_arr[i]-inc_arr[j])
    #Calculate the Gini coefficient
    g=abs_diff_sum/(2*avg_inc*(wt_arr.sum())**2)
    return g

#Generate a test income array
test_inc=np.random.choice(np.arange(1000)*10,size=100)

#Generate sample weights
test_wt=np.random.choice(np.arange(10,1000),size=len(test_inc))

#Compare Gini coefficients
print gini(test_inc)  
print gini_wt(test_inc) 

#Run with sample weights
print gini_wt(test_inc,test_wt)
```

That should be sufficient for our purposes, but do note that there also a pretty sweet covariance representation:

$$G = \frac{2}{\bar x} Cov(x,F(x))$$

This version can be extended to incorporate weights by modifying the cumulative function...

$$F(x_i) = \frac{i}{n} \rightarrow F(x_i) = \frac{w_i}{2} + \sum_{j=0}^{i-1} w_i$$

...and making sure to weight the incomes in the covariance calculation.  The full expression is as follows:

$$G = \frac{2}{\bar x} \sum_{i=1}^n w_i (x_i - \bar x) (\hat F(x_i) - \bar F)$$

where $\bar x$ is the weighted mean ($\sum_{i=1}^n w_i x_i$) and $\bar F$ is the weighted mean of $\hat F(x_i)$.  If more efficient algorithms are required that utilize sorted incomes, take note of the `np.argsort()` method, which returns the vector of indices that would sort an array.
