# Variance-Covariance: Geometric Intuition

This is just a quick notebook that helps to visualize why the determinant of a variance-covariance matrix can provide a singular value to describe the volume contained by the convex hull of a $n \times p$ matrix (a.k.a. a data set).

```python
import numpy as np
import pandas as pd
from pandas import Series,DataFrame
import scipy.stats as stats
from matplotlib.patches import Ellipse

%pylab inline

def simpleaxis(ax):
    ax.spines['top'].set_visible(False)
    ax.spines['right'].set_visible(False)
    ax.get_xaxis().tick_bottom()
    ax.get_yaxis().tick_left()
```

## Variance-Covariance Matrix

Variance is a measure of spread, and the sample version is generally represented by the following expression for a set of $n$ data points, $X$.

$$\frac{1}{n}\sum_i^n (x_i-E[x])^2$$

Covariance is a measure of association.  The sample covariance for $n$ realizations of a bivariate distribution of data ($X,Y$) is represented by the following expression.

$$\frac{1}{n}\sum_i^n (x_i-E[x])(y_i-E[y])$$

Beyond the clear mechanical parallel, both of these measures work in concert to describe the shape and orientation of the data contained in a matrix.  To see this, let us first consider the a bivariate distribution in which $X$ and $Y$ are uncorrelated.  In our example, $Y$ features $\sigma$=2 while $X$ has $\sigma$=1.  Both are centered on zero ($\mu_x=\mu_y=0$).

```python
#Define number of observations n
nobs=100000

#Generate uncorrelated series
x1=stats.norm.rvs(loc=0,scale=1,size=nobs)
y1=stats.norm.rvs(loc=0,scale=2,size=nobs)
uxy=np.column_stack((x1,y1))
print 'Uncorrelated Data'
print uxy
print uxy.shape

#Generate plot object
fig,ax=plt.subplots(figsize=(12,6))

#Plot overlapping histograms
ax.hist(uxy[:,0],ec='w',alpha=.5,color='b',bins=10,label='X')
ax.hist(uxy[:,1],ec='w',alpha=.5,color='r',bins=20,label='Y')

#Set labels
ax.set_title('$X$ and $Y$ Have Different Variances')
ax.set_xlabel('Realization Values of Random Variables $X$ and $Y$')
ax.set_ylabel('Frequency of Each Value')
ax.legend(loc='best')

#Fix axis
ax.set_xlim(-6,6)
simpleaxis(ax)
```

As can be seen, we have an $n \times p$ matrix where $n$=100,000 and $p$=2.  If we plot the joint distribution, we can see the shape of the data.  ([Hat tip](http://stackoverflow.com/questions/20126061/creating-a-confidence-ellipses-in-a-sccatterplot-using-matplotlib) on the ellipse code.)  

In the plot below, we can see the two dimensional data in space, each coordinate pair represented by a dot.  The ellipses capture the confidence bands at intervals of one standard deviation apiece.  For example, all of the dots within the first ellipse are within one standard deviation of the sample mean.  Since these data are normally distributed, that captures roughly 68% of all the dots in the plot.  The vectors emanating from the center are eigenvectors (scaled out to 3$\sigma$).  

```python
#Generate plot object
fig,ax=plt.subplots(figsize=(12,9))

def eigval(x,y):
    #Calculate variance-covariance matrix
    cov=np.cov(x,y)
    #Eigendecompose VC matrix
    lambda_,v=np.linalg.eig(cov)
    #Return square root of eigenvalues, and eigenvectors
    return (np.sqrt(lambda_),v)

def plot_ellipses(x,y,evals,evecs,n,ax):
    #Capture centroid
    exy=(x.mean(),y.mean())
    #For each of n confidence ellipses...
    for i in range(1,n+1):
        #...construct the ellipse...
        ell=Ellipse(xy=exy,
                    width=evals[0]*i*2,
                    height=evals[1]*i*2,
                    angle=np.rad2deg(np.arccos(evecs[0,0])))
        #...remove the internal color of the ellipse...
        ell.set_facecolor('none')
        ell.set_edgecolor('#707070')
        #...and plot the ellipse
        ax.add_artist(ell)
                    
def plot_data_shape(x,y,ax,n_ell=3):
    #Capture eigenvalues and eigenvectors
    evals,evecs=eigval(x,y)
    #Plot ellipses
    plot_ellipses(x,y,evals,evecs,n_ell,ax)
    #Plot data
    ax.scatter(x,y,c='#ff6600',lw=0,alpha=.2)
    #For each eigenvector...
    for e_, v_ in zip(evals, evecs.T):
        #...plot it out to the third std dev
        plt.plot([0, 3*e_*v_[0]], [0, 3*e_*v_[1]], color='#4D6A7B', lw=5)
    #Add axes
    ax.axvline(x=0,linestyle='--',linewidth=2,color='k')
    ax.axhline(y=0,linestyle='--',linewidth=2,color='k')
    #Fix range
    ax.set_xlim(-7,7)
    ax.set_ylim(-7,7)
     
#Plot data
plot_data_shape(uxy[:,0],uxy[:,1],ax)

#Add labels
ax.set_title('Joint Distribution of Uncorrelated $X$ and $Y$')
ax.set_xlabel('$X$')
ax.set_ylabel('$Y$');
```

Note that while the eigenvectors are very close to the coordinate axes, the differ by a hair.  This is only because of sampling variation.  We can expect this misalignment to decrease as $n$ goes to infinity.  In any event, this is what independent draws from $X$ and $Y$ look like.  Check out the variance-variance matrix.

$$\qquad {\boldsymbol {\Sigma }}_{X,Y}={\begin{bmatrix}{\boldsymbol {\sigma}}_{\mathit {X}}^2{\boldsymbol {\sigma }}_{\mathit {XY}}\\{\boldsymbol {\sigma }}_{\mathit {YX}}{\boldsymbol {\sigma}}_{\mathit {Y}}^2\end{bmatrix}}$$

```python
np.cov(uxy[:,0],uxy[:,1])
```

To reiterate, the variances (along the diagonal) are very close to those we defined for the underlying distribution. $\sigma_x$ and $\sigma_y$ are 1 and 2, respectively.  The covariances are not quite zero (due to sampling), but they are essentially negligible.

The question is, what happens when we insert some correlation in the underlying data generating process ($\sigma_{XY}$=0.7).

```python
#Define means for X and Y
means=[0,0]

#Define covariance matrix (note we are defining variances instead of std devs now)
cov_mat=[[1,.7],[.7,4]]

#Capture draws from dist
cxy=np.random.multivariate_normal(means,cov_mat,nobs)

print 'Correlated Data'
print cxy
print cxy.shape

#Generate plot object
fig,ax=plt.subplots(figsize=(12,6))

#Plot overlapping histograms
ax.hist(cxy[:,0],ec='w',alpha=.5,color='b',bins=10,label='X')
ax.hist(cxy[:,1],ec='w',alpha=.5,color='r',bins=20,label='Y')

#Set labels
ax.set_title('$X$ and $Y$ Still Have Different Variances')
ax.set_xlabel('Realization Values of Random Variables $X$ and $Y$')
ax.set_ylabel('Frequency of Each Value')
ax.legend(loc='best')

#Fix axis
ax.set_xlim(-6,6)
simpleaxis(ax)
```

Nothing but sampling has changed about the marginal distributions of $X$ or $Y$.  We have the same means and variances as in the uncorrelated case.  What does the shape of our data look like now?

```python
#Generate plot object
fig,ax=plt.subplots(figsize=(12,9))
     
#Plot data
plot_data_shape(cxy[:,0],cxy[:,1],ax)

#Add labels
ax.set_title('Joint Distribution of Correlated $X$ and $Y$')
ax.set_xlabel('$X$')
ax.set_ylabel('$Y$');
```

We have seen a shift!  Once we defined a positive covariance between $X$ and $Y$, the eigenvectors fell out of alignment with the coordinate axes.  Eigenvectors need not always be orthogonal to each other, but they will always be for symmetric square matrices.  As it turns out, variance-covariance matrices are both square and symmetric.  Consequently, eigenvectors can tell us about shifts in basis, and point to the farthest reaches of the data in question (that is, the directions in which the most variation occurs).  This ends up being the basis for why principal components can be identified.

So, we have seen that covariance can rotate the data in a given plane (or hyperplane in higher dimensions).  How does one affect the shape?  With the variance.  Let's hold covariance constant this time ($\sigma_{X,Y}$=0.7) and redefine the variances ($\sigma_X$=1 $\rightarrow \sigma_X$=1.5 & $\sigma_Y$=2 $\rightarrow \sigma_2$=0.5). 

```python
#Define means for X and Y
means=[0,0]

#Define covariance matrix (note we are defining variances instead of std devs now)
cov_mat2=[[1.5**2,.7],[.7,.5**2]]

#Capture draws from dist
cxy2=np.random.multivariate_normal(means,cov_mat2,nobs)

print 'Correlated Data'
print cxy2
print cxy2.shape

#Generate plot object
fig,ax=plt.subplots(figsize=(12,6))

#Plot overlapping histograms
ax.hist(cxy2[:,0],ec='w',alpha=.5,color='b',bins=30,label='X')
ax.hist(cxy2[:,1],ec='w',alpha=.5,color='r',bins=10,label='Y')

#Set labels
ax.set_title('We Have Changed the Variances for Both $X$ and $Y$')
ax.set_xlabel('Realization Values of Random Variables $X$ and $Y$')
ax.set_ylabel('Frequency of Each Value')
ax.legend(loc='best')

#Fix axis
ax.set_xlim(-6,6)
simpleaxis(ax)
```

How you like me now?

```python
#Generate plot object
fig,ax=plt.subplots(figsize=(12,9))
     
#Plot data
plot_data_shape(cxy2[:,0],cxy2[:,1],ax)

#Add labels
ax.set_title('Joint Distribution of Correlated $X$ and $Y$ w/ New Variances')
ax.set_xlabel('$X$')
ax.set_ylabel('$Y$');
```

It almost looks as though the ellipse has not only rotated, but turned within a third dimension.  To be clear, *this is not the case*.  We have just squeezed the data dramatically in the $Y$ dimension.  Note that the area seemed to remain relatively constant when the covariance was changed in isolation, but not when variances were changed in isolation.  This brings us to the point of this Notebook: determinants.

The covariance matrix helps us understand the shape of the data in all dimensions, but it can be a lot to take in.  Determinants can be a way of providing, essentially, a summary statistic.  Essentially, the determinant will [scale](http://stats.stackexchange.com/questions/176301/what-is-mean-by-single-numerical-measure-of-variability-in-multivariate-observat/176304#176304) with the variability inherent in a data set (by way of the covariance matrix).  Check out the determinants of the data we have simulated thus far.

```python
print 'Determinant of Uncorrelated Data:',np.linalg.det(np.cov(uxy[:,0],uxy[:,1]))
print 'Determinant of Correlated Data:',np.linalg.det(np.cov(cxy[:,0],cxy[:,1]))
print 'Determinant of Correlated Data w/ New Variances:',np.linalg.det(np.cov(cxy2[:,0],cxy2[:,1]))
```

Why does this happen?  Because the volume of an ellipsoid is [proportional](http://math.unt.edu/~tushar/project%20ideas/monthly%2096%20hannah%20geometric%20determinant.pdf) to the determinant of the matrix that defines it!
