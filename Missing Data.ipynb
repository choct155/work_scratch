# Filtering Missing Data

This is just a quick script demonstrating the filtering of tabular data with [pandas](http://pandas.pydata.org/).  Let's create some data.

```python
import pandas as pd
import numpy as np

d=pd.DataFrame(np.arange(100).reshape(25,4))

d
```

Now we need to generate some missing values.  We can just distribute them randomly.

```python
#Create an list of DF coordinates (housed in tuples) 
miss_index=zip(np.random.randint(0,25,20),np.random.randint(0,4,20))

#For each coordinate intersection...
for tup in miss_index:
    #...update the value to missing
    d.loc[tup]=np.NaN
    
d
```

Ok, let's say that we can deal with some missing values, but missing values in column `3` are just untenable and we want the affected rows out of our site.  The filter is a straightforward boolean mask based upon the reasonably solid handling of missing data by pandas.

```python
#Drop rows that have missing values in `3`
d[d[3].notnull()]
```

Let's say we want multiple conditions, because we do want to keep some of the rows that have missing values in `3`.  What if, for example, the vileness of the missing values is offset by column `0` containing a value that is a multiple of 32?

```python
#Filter null values in 3 or non-multiples of 32 in 0
d[(d[3].notnull())|(d[0]%32==0)] #where % is the modulus operator
```

As for [I/O](http://pandas.pydata.org/pandas-docs/dev/io.html), it's pretty clean to just read in the data unconditionally, and then filter as shown above.  That being said, if space is an issue, [chunk iterators](http://pandas.pydata.org/pandas-docs/dev/io.html#iterating-through-files-chunk-by-chunk) are available and, of course, low-level Python reads are always an option.
