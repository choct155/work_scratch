# Introduction to Gibbs Sampling

In the [Markov Chain Monte Carlo](https://en.wikipedia.org/wiki/Markov_chain_Monte_Carlo) world, there are a number of samplers used to conduct simulated draws from posterior distributions.  Some intuition behind how these work can be useful for understanding the results of Bayesian analyses.  This script is an effort to explore the mechanics of the Gibb sampler, but one could envision other samplers being captured here as well.

```python
import numpy as np
import pandas as pd
from pandas import Series, DataFrame
import scipy.stats as stats
import seaborn as sb

%pylab inline
def simpleaxis(ax):
    ax.spines['top'].set_visible(False)
    ax.spines['right'].set_visible(False)
    ax.get_xaxis().tick_bottom()
    ax.get_yaxis().tick_left()
    
plt.rcParams['axes.facecolor']='w'
plt.rcParams['axes.edgecolor']='k'
plt.rcParams['axes.linewidth']=1
```

## Gibbs - Basic Algorithm

The basic algorithm is outlined in [Bayesian Statistics and Marketing](http://abufara.com/abufara.net/images/abook_file/Bayesian%20Statistics%20and%20Marketing.pdf) (BSM by Rossi, Allenby, & McCulloch 2005).  It assumes a bivariate normal distribution, but it is generalizable to higher dimensions and a broader suite of distributions.

1. Start at $\theta_0$
2. Draw $\theta_1$
    + $\theta_{1,2} ~ N(\rho\theta_{0,1}, 1-\rho^2)$
    + $\theta_{1,1} ~ N(\rho\theta_{1,2}, 1-\rho^2)$
3. Repeat generalized version of Step 2:  Draw $\theta_{r+1}$
    + $\theta_{r+1,2} ~ N(\rho\theta_{r,1}, 1-\rho^2)$
    + $\theta_{1+1,1} ~ N(\rho\theta_{r+1,2}, 1-\rho^2)$
    
where $\rho$ is the correlation between residuals across random variables in the multivariate distribution.

## Gibbs - Simple Example

We can construct a simple example of this (also from BSM) by choosing our initial conditions and correlation parameter.

$\theta_0 = (\theta_{0,1},\theta_{0,2}) = (2,-2)$

$\rho = 0.9$

```python
theta0=(2,-2)
rho=0.9
```

Let us define two updating functions, one for each parameter.  These are just functional equivalents of the expressions in the basic algorithm above.

```python
def new_theta_2(rho,prev_theta_1):
    #Define mean and variance
    mean=rho*prev_theta_1
    variance=1-(rho**2)
    #Draw and return new value
    return stats.norm.rvs(mean,variance)

def new_theta_1(rho,curr_theta_2):
    #Define mean and variance
    mean=rho*curr_theta_2
    variance=1-(rho**2)
    #Draw and return new value
    return stats.norm.rvs(mean,variance)
```

Essentially, we are going to trace the path of this sampler from our initial `theta0` location until the location of the $n_{\text{th}}$ draw.

```python
def gibbs_path(n_draws,theta0=theta0):
    #Create a container to hold coordinates at each draw
    bvn_draw=[theta0]

    #Define number of draws
    n=n_draws

    #For each draw...
    for r in range(1,n+1):
        #...capture the new value for the second coordinate...
        tmp_2=new_theta_2(rho,bvn_draw[r-1][0])
        #...capture the new value for the first coordinate...
        tmp_1=new_theta_2(rho,tmp_2)
        #...augment list with theta values from draw r
        bvn_draw.append((tmp_1,tmp_2))

    #Capture draws in DF
    bvn_out=DataFrame({'theta1':[val[0] for val in bvn_draw],
                      'theta2':[val[1] for val in bvn_draw]})

    #Map colors based upon index position
    bvn_colors=[plt.cm.RdBu(val/float(n+1)) for val in range(n+1)]
#     print len(bvn_colors),len(bvn_out)
    bvn_out['color']=bvn_colors
    
    return bvn_out
    
#Capture gibbs path for n draws
bvn_df=gibbs_path(1000)

bvn_df.head()
```

The path of the draws gets a bit harder to see as the number of draws ($n$) increases.  Using color is not quite as precise, but a gives a general sense of the direction of movement.  In the plot below, the first draw is the deepest red.  The last one is the deepest blue.

```python
#Generate plot object
fig,ax=plt.subplots(figsize=(15,8))

#Plot draw coordinates
plt.scatter(x=bvn_df['theta1'],
            y=bvn_df['theta2'],
            c=bvn_df['color'],
            lw=0,
            s=100,
            alpha=.6)

#Fix labels
ax.set_xlabel('theta 1')
ax.set_ylabel('theta 2')
ax.set_title('Simulated Draws from Bivariate Normal (Gibbs)',fontsize=20)
ax.annotate('First Draw (theta0)',xy=(1.8,-1.8))

#Fix axes
simpleaxis(ax)
```

Joint histograms can help us get a sense of the shape of the simulated draws in two dimensions.

```python
#Plot joint density
sb.jointplot(x=bvn_df['theta1'],
             y=bvn_df['theta2'],
             size=10)
```

Notice the correlation between $\theta_1$ and $\theta_2$.  The simulated correlation is 0.86, and we know the actual correlation is $\rho$ = 0.9.  How quickly does our simulated correlation catch up with the actual value?

```python
#Create container for correlation values by sample size
bvn_corr=[]

#For 100,000 samples...
for i in range(0,2500,10):
    if i%500==0:
        print '>>Current Sample Size:',i
    #...capture the gibbs path with sample size equal to the index number...
    tmp=gibbs_path(i+1)
    #...capture the associated correlation (Pearson's)...
    tmp_corr=stats.pearsonr(tmp['theta1'],tmp['theta2'])[0]
    #...throw it in the container
    bvn_corr.append((i+1,tmp_corr))
    
#Capture correlations in DF
bvn_corr_df=DataFrame({'ndraws':[val[0] for val in bvn_corr],
                       'corr':[val[1] for val in bvn_corr]})

#Set index
bvn_corr_df.set_index('ndraws',inplace=True)
```

```python
#Generate plot object
fig,ax=plt.subplots(figsize=(15,6))

#Plot correlations
bvn_corr_df.plot(kind='line',ax=ax,lw=2,linestyle='-',color='#B0171F')

#Plot true correlation
ax.axhline(y=0.9,color='#A0A0A0',lw=3)

#Fix labels
ax.set_xlabel('Sample Size')
ax.set_ylabel('Correlation')
ax.set_title('Simulated Correlations between theta1 and theta2 (Gibbs)',fontsize=20)

#Fix axes
simpleaxis(ax)
```
