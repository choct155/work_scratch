# Mortgage Refinance Options

At the current time, a choice must be made regarding the refinancing of the house.  In general, it appears we can lower our out-of-pocket monthly costs, even if we switch to a 20-year loan (we currently have a 30-year fixed).  This Notebook explores tradeoffs.

```python
import numpy as np
import pandas as pd
from pandas import Series,DataFrame

pd.set_option('line_width',140)
```

## Amortization Schedule

The first step is to define a function that can provide a stream of payments with allocations for principal and interest by period.  It should be general enough to handle any loan term and interest rate.  The way we will implement this amortization function has two parts.  The first function captures the payment, and then this function is embedded in the larger amortization function that uses the payment.  In general, the payment function captures the following equation:

$PMT = p \times \frac{r}{1-(1+r)^{-t}}$  where

Variable|Definition|30-Year Value
--------|----------|-------------
$p$|Principal|\$380,000
$r$|Monthly Interest Rate| 4.0%
$t$|Term Length|360 months

The thirty year rate matches the Wells Fargo prelim estimate.  In general, the payment is not just $PMT=\frac{p}{t}$ because it must account for the interest.  What we require is the value of a constant payment that takes interest into account, which is another way of saying a constant payment that takes the time value of money (from the lender's perspective) into account.  To calculate this, it helps to understand the annuity concept.  This may sound initially like an unrelated topic, but annuity valuation is basically how we figure out the constant payment.  In fact, everything but $p$ on the right hand side of the $PMT$ equation is actually the inverse of the value of the applicable annuity.

### Annuity

An annuity is a payment that is the same nominal value in all periods for a given term.  So, if we received one dollar every month for five years, that would be considered an annuity.  Note that this example highlights that an annuity can be fit to any fixed period of time, not just years.  The term in this case would be 60 months.  

The question is, what is the value of a one dollar monthly annuity paid over five years?  We can start by thinking about the value of a single payment.  If this is the first month, the value is just one dollar because we are receiving the dollar now.  Said differently, the present value of the month 0 payment in month 0 is the nominal value: one dollar.  The value of the month 1 payment as viewed from month 0, however, must take into account the time value of money (which is operationalized in this case as the lender's interest rate).  The value of a payment of one dollar received a month from now is given by the following equation:

$pmt_1=\frac{1}{(1+r)}=(1+r)^{-1}$

*NOTE: Upper case $PMT$ signifies the value of the entire annuity, while lower case $pmt$ represents a single payment in the annuity stream of payments.*

---------------------------------------------

As a quick aside, here we are talking about constant nominal payments, but we can also consider the similarity of the constant value scenario.  If we had to pay someone an amount in month 1 that would be equivalent to receiving one dollar today, we would have to grow the nominal value by the interest rate.  Thus, the month 1 future value that corresponds to a present value of one dollar is given by the following:

$pmt_{FV_1}=(1+r)^1$

So, *when calculating future values* for a given present value, we multiply the present value by $(1+r)$.  When *discounting future payments* (that is, determining the present value), we divide by $(1+r)$.  I am stressing this point because it's the keystone.

---------------------------------------------------

In general, the value of one dollar received $n$ months from now is given by the following:

$pmt_n=\frac{1}{(1+r)^n}=(1+r)^{-n}$

Notice how this holds in month 0 as well.  $(1+r)^{-0}$ is just the nominal value of \$1.  The value of the annuity is simply the sum of these payments:

$Annuity=\sum_{i=0}^n pmt_i=\sum_{i=0}^n (1+r)^{-i}$

### Calculating Annuity Value

This all seems pretty straightforward.  Let's say we have a three period annuity paying one dollar per period at 10% interest.  In general, while we used month 0 (present) to illustrate the time value of money relationship, annuities are valued starting in month 1. 

$pmt_1=(1.1)^{-1}=0.9091$

$pmt_2=(1.1)^{-2}=0.8264$

$pmt_3=(1.1)^{-3}=0.7513$

The total annuity value is just their total:

$\sum_{i=0}^2 pmt_i = 2.4868$

Obviously this can easily be calculated with a spreadsheet, but it would be awfully convenient to not have to do so.  Ideally, we would have a method of calculation that provides a closed form (or formulaic) solution and is generalizable to provide the value of a one dollar annuity for any combination of interest rate, $r$, and term length, $t$.  To be concrete, we want some method of defining a function $A(r,t)$ that yields the value of the annuity.  In our toy example, we need $A(0.1,3)$ to yield $2.4868$.

The thing is, there is no easy way to add the payments together because the value distance between payments changes as the term goes on.  The value distance between $pmt_1$ and $pmt_2$ is $0.9091-0.8264=0.08264$ while the value distance between $pmt_2$ and $pmt_3$ is $0.8264-0.7513=0.07513$.  (Take note that the distance is just 10% of the later period value.)  Therefore, simple products or some derivative are out.

It turns out that we can use an indirect algebraic method to pull the answer out.  Essentially what we want to do is get ourselves an equation which we can use to solve for $A(r,t)$.  The strategy will be to compare two streams with a known difference.  Consider these two payment streams:

Period|Stream 1|Stream 2
------|--------|--------
1|<font color='blue'>$(1+r)^{-1}$</font>|$(1+r)^{-1} \times (1+r)=(1+r)^0=1$
2|<font color='red'>$(1+r)^{-2}$</font>|$(1+r)^{-2} \times (1+r)=$<font color='blue'>$(1+r)^{-1}$</font>
3|<font color='green'>$(1+r)^{-3}$</font>|$(1+r)^{-3} \times (1+r)=$<font color='red'>$(1+r)^{-2}$</font>
4|$(1+r)^{-4}$|$(1+r)^{-4} \times (1+r)=$<font color='green'>$(1+r)^{-3}$</font>
$\sum$|$A(r,4)$|$(1+r) \times A(r,4)$

As can be seen, all we did to get payments in Stream 2 was multiply the corresponding payment in Stream 1 by $(1+r)$.  *We do this so that if we subtract Stream 1 from Stream 2, most of the terms drop out.*  In particular, the color matched terms cancel, and we are left only with the payments in black.

Stream 1 less Stream 2: $(1+r) \times A(r,4) - A(r,4) = A(r,4) + rA(r,4) - A(r,4) = rA(r,4) = 1 - (1+r)^{-4}$

If we divide by $r$, we arrive at our value of the annuity...

$A(r,4) = \frac{1 - (1+r)^{-4}}{r}$

...which can be generalized to the standard equation:

$A(r,t) = \frac{1 - (1+r)^{-t}}{r}$

We can confirm that this works by checking against our toy example of a three period annuity at 10%.  Recall that when we simply added the payments up, we found that $\sum_{i=0}^2 pmt_i = 2.4868$.  If use our formula, we get the same result.

```python
(1-(1+.1)**(-3))/.1
```

If the formula looks familiar, it should!  We can recast our original payment equation:

$PMT = p \times \frac{r}{1-(1+r)^{-t}} = \frac{p}{A(r,t)}$

**Our constant payment in an amortization scheme is just the principal divided by a one dollar annuity consistent with the loan terms.**  To show that all of these concepts are tied together, consider a slight modification.  Someone has loaned us one dollar, and we must pay it back next month at 10% interest.  If we used the future value equation ($1 \times (1+0.1)^1$)...

```python
#Define principal
p=1

#Define interest rate
r=0.1

#Define period
per=1

#Calculate payment
p*(1+r)**per
```

...we owe them \$1.10 next month.  If we calculate with our new payment formula ($PMT = \frac{p}{A(r,t)}$)...

```python
#Define term limit
t=1

#Calculate payment
p*(r/(1-(1+r)**(-t)))
```

...we still owe them \$1.10. (Don't get hung up on the rounding, that's a function the way the computer stores numbers, not the formula.)  We can define a function to handle this calculation, though to be consistent with mortgage interest rates as commonly conveyed, we want it to take the [APR](http://en.wikipedia.org/wiki/Annual_percentage_rate) (annualized percentage rate) as an input instead of the monthly rate.  The conversion to monthly is done within the function.

```python
#Defioe function to calculate monthly payment
def payment(principal,rate,term):
    ratePerTwelve=rate/(12*100.)                                     #Define monthly interest rate
    result=principal*(ratePerTwelve/(1-(1+ratePerTwelve)**(-term)))  #Calculate payment
    return result

#Define principal
p=1

#Define rate (as percentage)
r=120.

#Define term length (in months)
t=1

print 'The payment for a 3-month $1 loan at 10% (120% APR):',payment(p,r,t)
```

### The Distribution of Principal Reduction and Interest

Given that we have used the annuity value to determine our constant payment, we know that each payment contains money that goes to principal reduction, and money that goes to interest.  The question is, how much goes to each function?  Full disclosure: I specifically chose the one period case to demonstrate the `payment()` function because all interest and principal must be paid at the same time.  If we went to two periods, only a portion of the principal would be repaid, *and this portion varies across periods*.  Therefore, in general, we cannot split the principal reduction evenly and pay the interest on the principal at the beginning of each month...

$\underbrace{\overbrace{\frac{p}{2}}^{Principal} + \overbrace{p \times (1+r) - p}^{Interest}}_{Month 1} + \underbrace{\overbrace{\frac{p}{2}}^{Principal} + \overbrace{\frac{p}{2} \times (1+r) - \frac{p}{2}}^{Interest}}_{Month 2} = \frac{p}{2} + pr + \frac{p}{2} + \frac{pr}{2} = p + \frac{3pr}{2} = p(1+\frac{3r}{2}) \neq 2 \times \frac{p}{A(r,2)}$

(*On the left hand side, the interest payments are calculated as remaining principal plus interest, and then the principal is taken out. It is crucial to see that we don't double pay interest.  In the second period we do not square the interest term $(1+r)$ because that would be like paying the first period's interest again for the remaining balance, on top of the second period's interest.*)

```python
print 'Equivalent principal reduction (EPR) in each period yields a total payment of',1*(1 + (3*.1)/2)
print 'Amortized payment (AP) stream yields a total payment of',2*payment(1,120.,2)
```

As we can see, in this case, going with equal principal reduction actually decreases the amount of total interest we have to pay.  Why does this happen?  It happens because we are paying more principal in period one than we would if we used the amortized scheme.  In all cases, the lender wants its return (that is, the interest income) first.  We then pay down the principal.  By adhering to equal principal payments, we have implicitly put a floor under how much of that principal we want to pay in period one.  In the amortization scheme, our first period payment is smaller because we pay less of the principal back.  The interest payment is the same in both cases: 10% of one dollar = \$0.10.

```python
print 'Period one principal repayment - EPR:',0.5
print 'Period one principal repayment - AP:',payment(1,120.,2)-.1
```

$\$2.10$

Why then, do we deal with amortization instead of using equivalent principal reduction?  Well, for one, financial institutions get more of a return this way.  Secondly, on the borrower side, constant payments make planning easier.  *Finally, and most importantly, if we do not gradually phase in principal repayment, the early payments can be large.*  When we are dealing with our little one dollar example, there is only a difference in first period payments of about $\$0.025$ cents.  On the scale that we care about, in the mortgage space, the gap grows.

If we are looking at a 30-year $\$380,000$ loan at 4%, the first principal payment is $\$547$ in the amortization scheme.  In the equivalent principal reduction scheme, that payment is $\$1,055$.  Since the interest payment is the same with both methods, the total first payment is $\$508$ higher.  

Of course, if one can absorb the early costs, there are some awfully nice payments on the back end.  We are now in a position to define an amortization function and show this comparitive action.

```python
def amort(principal,rate,term):
    '''Function returns an amortization schedule, given a principal amount, interest rate, and loan term'''
    #Define containers for table components
    begin=[]        #Remaining principal at the beginning of the period
    pay=[]          #Total payment amount (principal + interest)
    princ=[]        #Principal payment
    intr=[]         #Interest payment
    end=[]          #Remaining principal at the end of the period (after period payment)
    pay_no=[]       #Period number
    
    #Calculate payment with payment function
    pmt=payment(principal,rate,term)
    
    #Set the beginning balance for the first period
    begBal=principal
    
    #For each period in the term...
    for num in range(1,term+1):
        pay_no.append(num)                 #Capture the period number
        begin.append(begBal)               #Capture the remaining principal at the beginning of the period
        interest=begBal*(rate/(12*100.))   #Calculate the interest payment
        applied=pmt-interest               #Calculate the principal payment (total - interest)
        endBal=begBal-applied              #Calculate the remaining principal balance at the end of the period
        begBal=endBal                      #Set remaining balance to the starting balance for the next period
        
        #Capture all calculated figures in table component containers
        pay.append(pmt)
        princ.append(applied)
        intr.append(interest)
        end.append(endBal)
    
    #Capture containers fot table in a DataFrame
    df=DataFrame({'pay_no':pay_no,
                  'begin':begin,
                  'pmt':pay,
                  'principal':princ,
                  'interest':intr,
                  'end':end})
    return df

#Build amortization table
amtab=amort(380000,4.,360)

#Define new table for equivalent principal reduction approach
amtab_EPR=DataFrame({'pay_no':amtab['pay_no'],
                     'principal':amtab['begin'].iloc[0]/amtab['pay_no'].max()})

#Define beginning balance for first period
begin=[380000]
#For each subsequent period...
for i in range(len(amtab)-1):
    #...capture the new (after the first) balance less a principal payment
    begin.append(begin[i]-amtab_EPR['principal'].iloc[i])

#Throw new beginning balances in the DF
amtab_EPR['begin']=begin

#Calculate interest payments
amtab_EPR['interest']=amtab_EPR['begin'].apply(lambda x: x*(4./(12*100.)))

#Calculate payments
amtab_EPR['pmt']=amtab_EPR['interest']+amtab_EPR['principal']

#Generate plot object
fig,ax=plt.subplots(figsize=(16,5))

#Plot EPR payments
amtab_EPR['pmt'].plot(kind='line',ax=ax,lw=4,label='EPR',alpha=.7)

#Plot AP payments
amtab['pmt'].plot(kind='line',ax=ax,lw=4,color='red',label='AP',alpha=.7)

#Fix plot elements
ax.set_xlim([0,360])
ax.set_xlabel('Periods (months)')
ax.set_ylabel('Total Monthly Payment')
ax.set_title('Comparison of Total Payments - EPR vs AP')
ax.legend(loc='best');
```

What about the distribution of interest and principal in each payment?  Since there are only two values, we can just show the proportion allocated to principal reduction in both scenarios.

```python
#Define principal proportions
amtab['princ_pop']=amtab['principal']/amtab['pmt']
amtab_EPR['princ_pop']=amtab_EPR['principal']/amtab_EPR['pmt']

#Generate plot object
fig,ax=plt.subplots(figsize=(16,5))

#Plot EPR payments
amtab_EPR['princ_pop'].plot(kind='line',ax=ax,lw=4,label='EPR',alpha=.7)

#Plot AP payments
amtab['princ_pop'].plot(kind='line',ax=ax,lw=4,color='red',label='AP',alpha=.7)

#Fix plot elements
ax.set_xlim([0,360])
ax.set_xlabel('Periods (months)')
ax.set_ylabel('Total Monthly Payment')
ax.set_title('Comparison of Principal Proportion of Total Payment - EPR vs AP')
ax.legend(loc='best');
```

The curve is steeper for amortization because the ratio is changing due to interest payments getting smaller *and principal payments getting larger*.  In any event, now that we know how and why amortization works, we can think about how to compare the 20- and 30-year loan.

## How Do 20- and 30- Year Terms Compare?

We have two options at the current time for loans of \$380,000:

Option|Rate|Term
------|----|----
1|4.00%|360 months
2|3.75%|240 months

What is the cumulative total of the difference in payments?  This difference is the interest cost associated with going for the 30-year loan.

```python
#Capture amortization tables
amort1=amort(380000,4.,360)
amort2=amort(380000,3.75,240)

#Capture total difference
tot_pay_diff=amort1['pmt'].sum()-amort2['pmt'].sum()

print 'Total Increased Interest Due to 30-Year Term:',tot_pay_diff
print '30-Year Average Increased Interest Due to 30-Year Term:',tot_pay_diff/30
```

So, it looks like we would pay a bit more than $\$3,700$ in extra interest each year (on average) with the 30-year loan.  What do the cash flows look like?

```python
#Generate plot object
fig,ax=plt.subplots(figsize=(16,5))

#Plot EPR payments
amort2['pmt'].plot(kind='line',ax=ax,lw=4,label='20-Year',alpha=.7)

#Plot AP payments
amort1['pmt'].plot(kind='line',ax=ax,lw=4,color='red',label='30-Year',alpha=.7)

#Fix plot elements
ax.set_xlim([0,360])
ax.set_xlabel('Periods (months)')
ax.set_ylabel('Total Monthly Payment')
ax.set_title('Comparison of Total Payments - 20- vs 30-Year Term')
ax.legend(loc='best')
ax.annotate('Difference in Monthly Payment: '+str(amort2['pmt'].iloc[0]-amort1['pmt'].iloc[0]),xy=(50,2050))
```

It looks like going with the 30-year saves us about $\$440$ in monthly cash flow.  Over 20 years, that amounts to...

```python
print 'Total Cash Flow Difference Over 20 Years:',(amort2['pmt']-amort1['pmt']).iloc[0:241].sum()
```

So we lose $\$112,000$ in interest, but we gain an extra $\$105,000$ in cash flow over the first couple of decades.  If we were to compare the present value of $\$105,000$ in 20 years to $\$112,000$ in 30 years...

```python
print 'Present Value of $105,000 in 20 Years:',105000/(1+.031)**20
print 'Present Value of $112,000 in 30 Years:',112000/(1+.031)**30
```

...we would see that the present value of the cash flow difference is greater than the interest cost.  

It should be noted, that both of the tables were built using different interest rates, and I used neither to come up with these estimates.  First, we must use the same discount rate to put them on equal footing.  Second, we are concerned now with our time value of money, and not the interest rates charged by lenders.  I have defaulted to using OMB's nominal 20-year discount rate from [Circular A-94](https://www.whitehouse.gov/omb/circulars_a094/a94_appx-c).  It should be noted that there is considerable discussion about selection of appropriate discount rates, but this suffices for our purposes.  

In any event, I also mentioned before that we get a tax benefit from interest payments.  The benefit is equal to the product of the total interest paid in a year and our [marginal tax rate](http://taxfoundation.org/article/2015-tax-brackets).  Without getting too much into the weeds regarding legal changes, the likelihood of us staying in our current bracket, and our future tax liabilities, we can do a rough valuation by assuming we will be in the 33% bracket the whole time.  That means that we get about a third of that interest cost back.

```python
print 'Rough Valuation of Mortgage Interest Deduction Over Term:',tot_pay_diff*.33
```

Clearly, this makes the 30-year loan look even more attractive (on top of the present value gain).  One last argument has to do with fiscal stability, the cash flow give us not only the ability to invest elsewhere, but also the ability to respond to income shocks.  So, in the end, I think I lean towards the 30-year loan.
