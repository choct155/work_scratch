```python
import numpy as np
import pandas as pd
from pandas import Series, DataFrame
import seaborn as sb
import statsmodels.formula.api as smf

def simpleaxis(ax):
    ax.spines['top'].set_visible(False)
    ax.spines['right'].set_visible(False)
    ax.get_xaxis().tick_bottom()
    ax.get_yaxis().tick_left()

pd.set_option('display.width',140)
```

## Generating Synthetic Data

To keep all of this business abstract, we will use a synthetic data set, governed by the following model:

$$y = x_1 + x_2 + x_1*x_2$$

where the interaction $x_1*x_2$ is just the simple product of $x_1$ and $x_2$.  The data for the regressors $x_1$ and $x_2$ will be randomly generated (with $\mu=10$ & $\sigma=5$).  The data for the dependent $y$ will be constructed according to the above model, plus an additive noise component $\epsilon$ (with $\mu=0$ & $\sigma=2$).  Thus, the full spec is as follows:

$$y = x_1 + x_2 + x_1*x_2 + \epsilon$$

```python
#Generate values for x1 and x2
regressors=DataFrame(np.random.normal(10,5,size=(100,2)),columns=['x1','x2'])

#Generate interaction term
regressors['x1x2']=regressors['x1']*regressors['x2']

#Generate noise parameter
epsilon=DataFrame(np.random.normal(0,2,size=(100,1)),columns=['e'])

#Join right hand side of spec
data=regressors.join(epsilon)

#Construct dependent according to above model
data['y']=data['x1']+data['x2']+data['x1x2']+data['e']

#Reorder
data=data[['y','x1','x2','x1x2','e']]

data
```

## Regression

Now that we have our data, we can estimate the impact of $x_1$, $x_2$, and their interaction on $y$.

```python
#Construct spec
spec='y ~ x1 + x2 + x1x2'

#Instantiate model
mod=smf.ols(formula=spec,data=data)

#Fit model
results=mod.fit()

#Display summary results
results.summary()
```

As expected, we have very high fit and extremely clear signals from each model component.  Turns out it's easy when you create the data yourself.

## Plotting Interactions

If one wants to understand the impact of just $x_1$, she must wrestle with its entanglement with $x_2$ in the interaction term.  One way to do this is just set $x_2$ to its average value, and then plot the evolution of $y$ over all values of $x_1$.  Let's first do this manually:

1. Calculate the average value of $x_2$ as $x_{2,\sigma}$ (`x2avg`)
2. Calculate estimated values of the dependent ($\hat{y}$) with the following:
    $$\hat{y}=x_1 + x_{2,\sigma} + x_1*x_{2,\sigma} \; \forall \; x_1$$
3. Plot $\hat{y}$ vs $x_1$

```python
#Generate average of x2
data['x2avg']=data['x2'].mean()

#Calculate yhat
data['yhat']=data['x1']+data['x2avg']+data['x1']*data['x2avg']

#Generate plot object
fig,ax=plt.subplots(figsize=(10,10))

#Plot yhat vs x1
ax.scatter(data['yhat'],data['x1'],s=200,c='r',alpha=.5,lw=0)

#Fix labels
ax.set_xlabel('x1')
ax.set_ylabel('yhat')
ax.set_title('Marginal Effect of x1 on y')

#Fix axes
simpleaxis(ax)

#Fix background
ax.patch.set_facecolor('w')

data.head()
```

This plot shows that the marginal effect of $x_1$ on $y$ remains consistently positive throughout the range of $x_1$ values.

If we want a more continuous measure that captures the marginal effects of both $x_1$ and $x_2$, we can employ stock methods designed to capture changes in the dependent in the background as gradation of colors.

```python
sb.interactplot('x1','x2','y',data,cmap='coolwarm',filled=True,levels=25)
```

This plot shows that high values of $x_1$ and $x_2$ are associated with higher values of $y$ (indicated by the <font color='red'>red</font> background).  High values of $x_1$ or $x_2$ in isolation cannot match the combined effect.  Furthermore, and this highlight the value of interactions, when we start with high values of $x_1$, increases in $x_2$ push up $y$ faster than if we start with low values of $x_1$. 

## Exploring Interactions with More Regressors

In general, whenever we want to understand the marginal impact of a single variable (e.g. $x_1$), we must make structural impositions on the behavior of the remaining variables (e.g. $x_2$) so that we can systemically extract said marginal impact.  In our case, the imposition we have made is to set $x_2$ to its sample mean.  There are two reasons this is useful.  First, and foremost, we have held it constant at a fixed value.  Therefore, we have eliminated a source of variation in the dependent so that we could focus on $x_1$.  

Second, the selection of the mean is useful.  Technically, we could have chosen any point in the distribution, so why choose the average?  In most cases, a regression model reveals the average effect of the regressors included in the specification.  In reality, the effects may differ at different values of the regressor in question. We focus on the average because it is analytically convenient, and often the quantity of interest anyway.  In any event, because regression models typically yield average effects, we want some way of rendering a regressor inert without biasing our results.  If we set a regressor to its average value, it removes variation in the regressor without losing said average effect.

In general, this approach will still work with more than two variables.  To isolate the effect of a given regressor, set all of the others to their conditional expectation (i.e. their sample average).  I will concede that this is not the tightest description in the world, but I think it captures the intuition well enough.
