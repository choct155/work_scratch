```python
import numpy as np
import pandas as pd
from pandas import Series, DataFrame
```

```python
data=np.random.uniform(size=100)*100

data
```

```python
weights=np.random.randint(low=1000,high=5000,size=100)

weights
```

```python
test=DataFrame({'inc':data,
                'wt':weights})

test
```

```python
#Sort the data on income
test=test.sort(columns='inc')

#Create cumulative sum variable
test['cwt']=test['wt'].cumsum()

#Create cumulative sum as percentage of total weight
test['cwt_prop']=test['cwt']/test['wt'].sum()

print test.to_string()
```

```python
test[test['cwt_prop']>.5].iloc[0]
```
