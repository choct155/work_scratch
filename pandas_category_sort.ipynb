```python
import numpy as np
import pandas as pd
from pandas import Series, DataFrame

d=DataFrame(np.random.uniform(size=100).reshape(25,4))

d
```

```python
cat_list=['first','second','third','fourth']

d['cat']=np.random.choice(cat_list,size=len(d))

d.sort('cat')
```

```python
d['cat_sort']=pd.Categorical(d['cat'],
                            categories=cat_list,
                            ordered=True)

d.sort('cat_sort')
```
